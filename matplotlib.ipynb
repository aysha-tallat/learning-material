# Object Oriented apporach

```python
fig,ax=plt.subplots(figsize=(6,5))
over40.plot(kind='scatter',
           x='Age',
           y='Cholesterol',
           c='HeartDisease')
```

# Life Cycle of OO Method 

```python
# 1-create a figure and graph area
fig,ax=plt.subplots(figsize=(10,6));
#2-plot the data
scatter_plot=ax.scatter(x=heart['Age'],
                       y=heart['Cholesterol'],
                       c=heart['HeartDisease']);
#3-customization
ax.set(title='Suspectability of Heart Diseases',
      xlabel='Age',
       ylabel='Cholesterol');
#4-set the legends
ax.legend(scatter_plot.legend_elements(),title='target');
scatter_plot;
```

# Life Cycle of OOmethod Advanced

```python
#1-create plot area
fig,(plot1,plot2)=plt.subplots(nrows=2,ncols=1,figsize=(10,10))
#2-add data to plot
scatter_plot=plot1.scatter(x=over40['Age'],
                         y=over40['Cholesterol'],
                         c=over40['HeartDisease'])
scatter_plot2=plot2.scatter(x=over40['Age'],
                         y=over40['RestingBP'],
                         c=over40['HeartDisease'])
#3-customization
plot1.set(title='Heart Disease Problem',
         xlabel='Age',
         ylabel='Cholesterol')
plot2.set(title='Heart Disease Problem',
         xlabel='Age',
         ylabel='RestingBP')
#4-set legends
plot1.legend(*scatter_plot.legend_elements(),
         title='survived or not')
plot2.legend(*scatter_plot2.legend_elements(),
         title='survived or not')
```

```python
#1-create plot area
fig,(plot1,plot2)=plt.subplots(nrows=2,ncols=1,figsize=(10,10),sharex=True)
#2-add data to plot
scatter_plot=plot1.scatter(x=over40['Age'],
                         y=over40['Cholesterol'],
                         c=over40['HeartDisease'])
scatter_plot2=plot2.scatter(x=over40['Age'],
                         y=over40['RestingBP'],
                         c=over40['HeartDisease'])
#3-customization
plot1.set(title='Heart Disease Problem',
         xlabel='Age',
         ylabel='Cholesterol')
plot2.set(title='Heart Disease Problem',
         xlabel='Age',
         ylabel='RestingBP')
#4-set legends
plot1.legend(*scatter_plot.legend_elements(),
         title='survived or not')
plot2.legend(*scatter_plot2.legend_elements(),
         title='survived or not')
```

```python
#1-create plot area
fig,(plotm,plotf)=plt.subplots(nrows=2,ncols=1,figsize=(10,10),sharex=True)
#2-add data to the plot
maleplot=plotm.scatter(x=male['Age'],
                      y=male['Cholesterol'],
                      c=male['HeartDisease'])
femaleplot=plotf.scatter(x=female['Age'],
                      y=female['Cholesterol'],
                      c=female['HeartDisease'])
#3=Customization
plotm.set(title='Heart Disease Problem',xlabel='Age',ylabel='Cholesterol')
plotf.set(xlabel='Age',ylabel='Cholesterol')
#4-set legends
plotm.legend(*maleplot.legend_elements(),title='survived or not(male)')
plotf.legend(*femaleplot.legend_elements(),title='Survived or Not(female)')
```

```python
#1-create plot area
fig,(plotm,plotf)=plt.subplots(nrows=2,ncols=1,figsize=(10,10),sharex=True)
#2-add data to the plot
maleplot=plotm.scatter(x=male['Age'],
                      y=male['Cholesterol'],
                      c=male['HeartDisease'])
femaleplot=plotf.scatter(x=female['Age'],
                      y=female['Cholesterol'],
                      c=female['HeartDisease'])
#3=Customization
plotm.set(title='Heart Disease Problem',xlabel='Age',ylabel='Cholesterol')
plotf.set(xlabel='Age',ylabel='Cholesterol')
#4-set legends
plotm.legend(*maleplot.legend_elements(),title='survived or not(male)')
plotf.legend(*femaleplot.legend_elements(),title='Survived or Not(female)')
#5-darw meanline
plotm.axhline(y=male['Cholesterol'].mean(),linestyle='-')
plotf.axhline(y=female['Cholesterol'].mean(),linestyle='--')
```

```python
#1-create plot area
fig,(plotm,plotf)=plt.subplots(nrows=2,ncols=1,figsize=(10,10),sharex=True)
#2-add data to the plot
maleplot=plotm.scatter(x=male['Age'],
                      y=male['Cholesterol'],
                      c=male['HeartDisease'],
                      cmap='RdBu')
femaleplot=plotf.scatter(x=female['Age'],
                      y=female['Cholesterol'],
                      c=female['HeartDisease'],
                        cmap='coolwarm')
#3=Customization
plotm.set(title='Heart Disease Problem',xlabel='Age',ylabel='Cholesterol')
plotf.set(xlabel='Age',ylabel='Cholesterol')
#4-set legends
plotm.legend(*maleplot.legend_elements(),title='survived or not(male)')
plotf.legend(*femaleplot.legend_elements(),title='Survived or Not(female)')
#5-darw meanline
plotm.axhline(y=male['Cholesterol'].mean(),linestyle='-')
plotf.axhline(y=female['Cholesterol'].mean(),linestyle='--')
#6-add or change colour of ploted data via cmap
```
